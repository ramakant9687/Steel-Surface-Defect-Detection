```python
dirs=p.glob("*")
for d in dirs:
    print(d)
```

```python
image_data=[]
labels=[]
labels_dict={"Crazing":0,"Inclusion":1,"Patches":2,"PittedSurface":3,"RolledInScale":4,"Scratches":5}
label_to_defect={0:"Crazing",1:"Inclusion",2:"Patches",3:"PittedSurface",4:"RolledInScale",5:"Scratches"}
```

```python
for folder_dir in dirs:
    label=(str(folder_dir).split("\\")[-1])
    cnt=0
    print(folder_dir)
    
    
    #Iterate over folder_dir and pick all images of the pokemon
    for img_path in folder_dir.glob("*jpg"):
        #print(img_path)
        img=image.load_img(img_path,target_size=(100,100))
        img_array=image.img_to_array(img)
        image_data.append(img_array)
        labels.append(labels_dict[label])
        cnt+=1
    print(cnt)
        
```
